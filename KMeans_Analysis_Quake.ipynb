```python
import pandas as pd
```

```python
dataset = pd.read_csv('quake.dat', sep=',', skiprows=7, engine='python', names=['Focal_depth', 'Latitude', 'Longitude', 'Ritcher'])
dataset
```

```python
from sklearn.cluster import KMeans
kmeans_instances = [KMeans(n_clusters=i).fit(dataset) for i in range(1,11)]
```

Elbow method

```python
import numpy as np
from scipy.spatial.distance import cdist

X = dataset
distortions = []
inertias = []
mapping1 = {}
mapping2 = {}

for i in range(1,11):

  distortions.append(sum(np.min(cdist(X, kmeans_instances[i-1].cluster_centers_,
                                        'euclidean'), axis=1)) / X.shape[0])
  inertias.append(kmeans_instances[i-1].inertia_)
 
  mapping1[i-1] = sum(np.min(cdist(X, kmeans_instances[i-1].cluster_centers_,
                                   'euclidean'), axis=1)) / X.shape[0]
  mapping2[i-1] = kmeans_instances[i-1].inertia_
```

```python
for key, val in mapping1.items():
    print(f'{key} : {val}')
```

```python
from matplotlib import pyplot as plt

plt.plot(range(1,11), distortions, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.savefig("kmeans_quake_distortion.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_distortion.pdf", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_distortion.svg", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```

```python
for key, val in mapping2.items():
	print(f'{key} : {val}')
```

```python
plt.plot(range(1,11), inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.savefig("kmeans_quake_inertia.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_inertia.pdf", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_inertia.svg", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```

Davies Bouldin score

```python
from sklearn.metrics import davies_bouldin_score

davies_scores = []
for key, kmeans_instance in enumerate(kmeans_instances):
  if key >= 2: 
    davies_scores.append(davies_bouldin_score(X, kmeans_instance.labels_))
    print(key,':',davies_scores[-1])
```

```python
plt.plot(range(2,10), np.array(davies_scores), 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Davies Bouldin Score')
plt.title('Davies Bouldin Score for K clusters')
plt.savefig("kmeans_quake_davies_bouldin.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_davies_bouldin.pdf", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_davies_bouldin.svg", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```

Calinski Harabasz score

```python
from sklearn.metrics import calinski_harabasz_score

calinski_scores = []
for key, kmeans_instance in enumerate(kmeans_instances):
    if key >= 2: 
        calinski_scores.append(calinski_harabasz_score(X, kmeans_instance.labels_))
        print(key,':',calinski_scores[-1])
```

```python
plt.plot(range(2,10), calinski_scores, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Calinski Harabasz Score')
plt.title('Calinski Harabasz Score for K clusters')
plt.savefig("kmeans_quake_calinski_harabasz.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_calinski_harabasz.pdf", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_calinski_harabasz.svg", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```

Silhouette score

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for key, kmeans_instance in enumerate(kmeans_instances):
    if key >= 2: 
        silhouette_scores.append(silhouette_score(X, kmeans_instance.labels_, metric='euclidean'))
        print(key,':',silhouette_scores[-1])
```

```python
plt.plot(range(2,10), silhouette_scores, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Coefficient for K clusters')
plt.savefig("kmeans_quake_silhouette.png", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_silhouette.pdf", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.savefig("kmeans_quake_silhouette.svg", transparent = True, bbox_inches = 'tight', pad_inches = 0)
plt.show()
```
